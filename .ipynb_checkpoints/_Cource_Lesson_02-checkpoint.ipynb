# 2. Взаимодействия
## 2.1. Количественные с колличественными

```python
feat = [f for f in df.columns if 'charge' in f] # Находим столбцы, в названиях которых есть слово 'charge'
feat
```

```python
df[feat].hist(); # Строим гистограммы для всех отобранных признаков
```

```python
# Строим Скетер-плот
sns.pairplot(df[feat]);
```

```python
# Раскрашиваем данные
df['Churn'].map({False: 'blue', True: 'orange'})
```

```python
# Строим отношение вечерних и международных звонков с раскраской 
plt.scatter(df[df['Churn']]['Total eve charge'], df[df['Churn']]['Total intl charge'],
           color='orange', label='Ушедшие');

plt.scatter(df[~df['Churn']]['Total eve charge'], df[~df['Churn']]['Total intl charge'],
           color='blue', label='Лояльные');

plt.xlabel('Вечерние начисления');
plt.ylabel('Международные начисления');
plt.title('Распределение');
plt.legend();
```

```python
# Корреляция (Корреляция Пирсона)
df.corr()
```

```python
# Раскрашиваем матрицу seaborn.heatmap() - тепловая карта
sns.heatmap(df.corr());
```

```python
# На тепловой карте видно, что корреляция некоторых данных практически равна 1. Можно удалить эту информацию
# Вспоминаем созданный нами список столбцов - их будем удалять
feat
```

```python
# Удаление данных из датафрэйма. Особое внимение уделить осям
# drop() по умолчанию удаляет строки - необходимо задать ось для удаления
# axis = 0 - строки
# axis = 1 - столбцы
# drop() не изменяет исходный датафрэйм
# inplace=True - вносит изменение в исходный датафрэйм
df.drop(feat, axis=1, inplace=True) # ВНИМАНИЕ - ИСХОДНИК ИЗМЕНЕН
```

```python
df.columns
```

```python
sns.heatmap(df.corr());
```

## 2.2. Количественный-категориальные и количественный-бинарный признаки

```python
# Ящики с усами
# Зависимость оттока от количества минут в день
# На графике видно, что медианы четко отличаются - те, кто ушел в среднем говорят по телефону БОЛЬШЕ
#
sns.boxenplot(x='Churn', y='Total day minutes', data=df);
```

## 2.3. Категориальный с категориальным признаки

```python
# Зависимость оттока от подключенного международного плана (сводная таблица)
pd.crosstab(df['Churn'], df['International plan'])
```

```python
# hue - столбец для раскрашивания.
# Определяем, как изменяется отток в случае подключения международного роуминга
sns.countplot(x='International plan', hue='Churn', data=df);
```

```python
# Зависимость оттока от звонков в сервисный центр
sns.countplot(x='Customer service calls', hue='Churn', data=df);
```

# t-SNE - обучение многообразию (представление)

```python
from sklearn.manifold import TSNE
```

```python
# Создаем объект TSNE
tsne = TSNE(random_state=0)
```

```python
# Преобразуем исходный датафрэйм
df2 = df.drop('State', axis=1)
```

```python
df2['International plan'] = df['International plan'].map({'Yes': 1, 'No': 0})
df2['Voice mail plan'] = df['Voice mail plan'].map({'Yes': 1, 'No': 0})
```

```python
df2.head()
```

```python
%%time
tsne.fit(df2)
```

```python
tsne.embedding_.shape
```

```python
plt.scatter(tsne.embedding_[:, 0], tsne.embedding_[:, 1]);
```

```python
plt.scatter(tsne.embedding_[df2['Churn'].values, 0], 
            tsne.embedding_[df2['Churn'].values, 1], color='orange'
           );

plt.scatter(tsne.embedding_[~df2['Churn'].values, 0], 
            tsne.embedding_[~df2['Churn'].values, 1], color='blue'
           );
```
